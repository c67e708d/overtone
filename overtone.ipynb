```python
import pandas as pd
import numpy as np
```

```python
n = 60
```

```python
equal_temperament = np.power(2, (np.arange(12) ) / 12)
```

```python
key_signature_name_list = [
    "C",
    "C sharp",
    "D",
    "D sharp",
    "E",
    "F",
    "F sharp",
    "G",
    "G sharp",
    "A",
    "A sharp",
    "B",
]
```

```python
def find_nearest_equal_temperament(value):
    array = equal_temperament
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
def find_nearest_key(value):
    array = np.array(equal_temperament)
    idx = (np.abs(array - value)).argmin()
    return key_signature_name_list[idx]
```

```python
def normalize(frequency):
    f = frequency
    while f >= 2.0:
        f /= 2.0
    return f
```

```python
def octave(frequency):
    return int(np.log2(frequency))
```

```python
overtone = np.arange(n) + 1
```

```python
df = pd.DataFrame(dict(overtone=overtone))
df["normalized_overtone"] = df.overtone.apply(normalize)
df["nearest_equal_temperament"] = df.normalized_overtone.apply(
    find_nearest_equal_temperament
)
df["overtone_par_equal_temperament"] = (
    df.normalized_overtone / df.nearest_equal_temperament
)
df["key"] = df.normalized_overtone.apply(find_nearest_key)
df["cent"] = df.overtone_par_equal_temperament.apply(lambda x: 1200 * np.log2(x))
df["octave"] = df.overtone.apply(octave)
```
